# 3. Data Preparation

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('loan_data_2007_2014.csv', index_col = 0)
df.sample(3, random_state = 42)
```

```python
list_item = []
for col in df.columns :
    list_item.append([col, df[col].dtype, df[col].count(), df[col].isna().sum(),
                      100 * df[col].isna().sum() / len(df[col]),
                      df[col].min() if df[col].dtype in ['int64', 'float64'] else '-',
                      df[col].max() if df[col].dtype in ['int64', 'float64'] else '-',
                      df[col].nunique(), df[col].unique()])
desc_df = pd.DataFrame(data = list_item, columns = ['feature', 'data_type', 'fill_num', 'null_num', 'null_percent', 'min_value', 'max_value', 'unique_num', 'unique_sample'])
desc_df
```

```python
# Definisikan kolom-kolom yang akan ditangani missing values
df_uniq = ['id', 'member_id', 'url', 'policy_code', 'application_type']
df_null = desc_df[desc_df.null_percent > 20].feature.tolist()
df_rmve = desc_df[(desc_df.null_percent <= 20) & (desc_df.null_percent > 0)].feature.tolist()
```

```python
# Hapus kolom dengan jumlah nilai unique hanya 1
for col in df_uniq :
    df = df.drop(col, axis = 1)
```

```python
# Hapus kolom dengan nilai null lebih dari 20% dari total data
for col in df_null :
    df = df.drop(col, axis = 1)
```

```python
# Hapus baris dengan nilai null kurang dari 20% dari total data
df = df.dropna(subset = df_rmve)
```

```python
df_nums = df.select_dtypes(include = ['int64', 'float64'])
df_cats = df.select_dtypes(include = 'object')

nums = df_nums.columns.tolist()
cats = df_cats.columns.tolist()
```

```python
# Buat fungsi untuk pengecekan outlier
def check_pct_outlier(data) :
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers   = data[(data < lower_bound) | (data > upper_bound)]
    n_outliers = data[(data >= lower_bound) & (data <= upper_bound)]
    l_outliers = data[(data < lower_bound)]
    u_outliers = data[(data > upper_bound)]
    
    outlier_percentage   = (len(outliers) / len(data)) * 100
    n_outlier_percentage = (len(n_outliers) / len(data)) * 100
    l_outlier_percentage = (len(l_outliers) / len(data)) * 100
    u_outlier_percentage = (len(u_outliers) / len(data)) * 100
    
    return outlier_percentage, n_outlier_percentage, l_outlier_percentage, u_outlier_percentage
```

```python
list_item = []
for col in df[nums].columns :
    list_item.append([col, df[nums][col].dtype, df[nums][col].count(), df[nums][col].isna().sum(),
                      100 * df[nums][col].isna().sum() / len(df[nums][col]),
                      df[nums][col].min() if df[nums][col].dtype in ['int64', 'float64'] else '-',
                      df[nums][col].max() if df[nums][col].dtype in ['int64', 'float64'] else '-',
                      df[nums][col].nunique(), df[nums][col].unique(),
                      check_pct_outlier(df[nums][col])[0], check_pct_outlier(df[nums][col])[1]])
desc_df = pd.DataFrame(data = list_item, columns = ['feature', 'data_type', 'fill_num', 'null_num', 'null_percent', 'min_value', 'max_value', 'unique_num', 'unique_sample', 'out_percent', 'no_out_percent'])
desc_df.sort_values('out_percent')
```

```python
# Buatkan log transformation
for col in nums :
    df[col] = np.log1p(df[col])
```

```python
# Penentuan label
def risk_category(x) :
    if x in ['Fully Paid'] :
        return 0
    elif x in ['Current'] :
        return 1
    elif x in ['In Grace Period', 'Charged Off', 'Late (16-30 days)', 'Late (31-120 days)', 'Default'] :
        return 2

df['loan_status'] = df['loan_status'].apply(lambda x : risk_category(x))
```

```python
# Buatkan label encoding
from sklearn.preprocessing import LabelEncoder

for col in cats :
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
# Penentuan feature selection berdasarkan multikolinearitas kolom
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data['feature'] = df[nums].columns
vif_data['VIF']     = [variance_inflation_factor(df[nums].values, i) for i in range(len(df[nums].columns))]

vif_data.sort_values('VIF', ascending = False).T
```

```python
corr_remove = ['funded_amnt', 'funded_amnt_inv', 'installment', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv', 'total_rec_int',
               'total_rec_prncp', 'total_acc', 'collection_recovery_fee', 'total_rev_hi_lim', 'annual_inc', 'revol_bal']

for i in corr_remove :
    df.drop(i, axis = 1, inplace = True)
    nums.remove(i)
```

```python
X = df.drop(columns = ['loan_status'])
y = df['loan_status']

y.value_counts(normalize = True)
```

```python
# Pemisahan data train dan test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Impor data
X_train.to_csv('feature_train.csv', index = False)
X_test.to_csv('feature_test.csv', index = False)
y_train.to_csv('label_train.csv', index = False)
y_test.to_csv('label_test.csv', index = False)
```
