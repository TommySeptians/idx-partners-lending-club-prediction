```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('loan_data_2007_2014.csv', index_col = 0)
df.sample(3, random_state = 42)
```

```python
df.shape
```

```python
list_item = []
for col in df.columns :
    list_item.append([col, df[col].dtype, df[col].count(), df[col].isna().sum(),
                      100 * df[col].isna().sum() / len(df[col]),
                      df[col].min() if df[col].dtype in ['int64', 'float64'] else '-',
                      df[col].max() if df[col].dtype in ['int64', 'float64'] else '-',
                      df[col].nunique(), df[col].unique()])
desc_df = pd.DataFrame(data = list_item, columns = ['feature', 'data_type', 'fill_num', 'null_num', 'null_percent', 'min_value', 'max_value', 'unique_num', 'unique_sample'])
desc_df
```

```python
desc_df.data_type.value_counts()
```

```python
desc_df[desc_df.null_percent > 20].shape[0]
```

```python
desc_df[(desc_df.null_percent <= 20) & (desc_df.null_percent > 0)].shape[0]
```

### Beberapa analisa dari dataset yang ditemukan antara lain :
- Terdapat 466285 baris dan 74 kolom pada dataset ini.
- Ada 22 kolom kategorikal dan 52 kolom numerikal.
- Beberapa kolom memiliki nilai yang *unique*, antara lain `id`, `member_id` dan `url`. Untuk kolom-kolom tersebut perlu dilakukan *drop column*.
- Kolom `policy_code` dan `application_type` memiliki jumlah *unique* hanya satu, sehingga perlu dilakukan *drop column*.
- Ada 22 kolom yang memiliki data kosong lebih dari 20 persen dari keseluruhan data, lebih baik dilakukan *drop column* juga.
- Ada 18 kolom yang memiliki data kosong di bawah 20 persen dari keseluruhan data, untuk ini dilakukan *remove missing values* tersebut.
